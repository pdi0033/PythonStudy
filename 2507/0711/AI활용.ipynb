```python
import random

def updown_game():
    print("üéØ ÏóÖÎã§Ïö¥ Í≤åÏûÑÏùÑ ÏãúÏûëÌï©ÎãàÎã§!")
    print("1Î∂ÄÌÑ∞ 100 ÏÇ¨Ïù¥Ïùò Ïà´Ïûê Ï§ë ÌïòÎÇòÎ•º ÎßûÏ∂∞Î≥¥ÏÑ∏Ïöî.")

    answer = random.randint(1, 100)
    attempts = 0

    while True:
        try:
            guess = int(input("Ïà´ÏûêÎ•º ÏûÖÎ†•ÌïòÏÑ∏Ïöî: "))
            attempts += 1

            if guess < 1 or guess > 100:
                print("‚ö†Ô∏è 1Î∂ÄÌÑ∞ 100 ÏÇ¨Ïù¥Ïùò Ïà´ÏûêÎ•º ÏûÖÎ†•Ìï¥Ï£ºÏÑ∏Ïöî.")
                continue

            if guess < answer:
                print("üîº UP!")
            elif guess > answer:
                print("üîΩ DOWN!")
            else:
                print(f"üéâ Ï†ïÎãµÏûÖÎãàÎã§! ÏãúÎèÑÌïú ÌöüÏàò: {attempts}Ìöå")
                break

        except ValueError:
            print("‚ùå Ïà´ÏûêÎßå ÏûÖÎ†•Ìï¥Ï£ºÏÑ∏Ïöî.")

# Í≤åÏûÑ Ïã§Ìñâ
updown_game()
```

```python
from deepface import DeepFace
import matplotlib.pyplot as plt
import cv2
import matplotlib
import matplotlib.pyplot as plt

# ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï (Windows Í∏∞Î≥∏ Ìè∞Ìä∏)
matplotlib.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False  

# 1. Ïù¥ÎØ∏ÏßÄ Í≤ΩÎ°ú ÏûÖÎ†• (ÏßÅÏ†ë ÏóÖÎ°úÎìúÌïòÍ±∞ÎÇò Î°úÏª¨ Í≤ΩÎ°ú ÏÇ¨Ïö©)
img_path = "face1.jpg"  # Ïòà: 'C:/Users/you/Pictures/face.jpg'

# 2. Ïù¥ÎØ∏ÏßÄ Î°úÎî© Î∞è ÌëúÏãú
img = cv2.imread(img_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
plt.title("Î∂ÑÏÑù ÎåÄÏÉÅ Ïù¥ÎØ∏ÏßÄ")
plt.show()

# 3. Í∞êÏ†ï Î∂ÑÏÑù ÏàòÌñâ
result = DeepFace.analyze(img_path=img_path, actions=['emotion'])

# 4. Í≤∞Í≥º Ï∂úÎ†•
emotion_labels = {
    'angry': 'ÌôîÎÇ®',
    'disgust': 'ÌòêÏò§',
    'fear': 'ÎëêÎ†§ÏõÄ',
    'happy': 'ÌñâÎ≥µ',
    'sad': 'Ïä¨Ìîî',
    'surprise': 'ÎÜÄÎûå',
    'neutral': 'Ï§ëÎ¶Ω'
}

print("üòÄ Í∞êÏ†ï Î∂ÑÏÑù Í≤∞Í≥º:")
for eng, score in result[0]['emotion'].items():
    kor = emotion_labels.get(eng, eng)
    print(f"{kor}: {score:.2f}%")
```
