## 챗봇의 채팅창 필터 -> '프롬프트 템플릿'

```python
customer_email = '''
아니 어제 산 모니터가 안켜짐. 아게 맞음?
반품하려고 했더니 고객센터도 전화를 안받네
장사 접어라
빡치게 만들지 말고;;
'''
```

```python
style = '정중하고 상냥한 문의. 비속어와 은어, 사회적 교양에 맞지 않는 말은 생략할 것'
```

```python
prompt = f'''
    주어진 텍스트를 {style}을 적용하여 변환해줘.
    주어진 텍스트 : {customer_email}
'''
```

```python
# 랭체인의 다양한 모델 중, openai 와으이 통신을 수행할 것
from langchain.chat_models import ChatOpenAI

# 채팅을 위한 프롬프트 양식을 만듦
from langchain.prompts import ChatPromptTemplate
```

```python
prompt = '''
    주어진 텍스트를 {style}을 적용하여 변환해줘.
    주어진 텍스트 : {customer_email}
'''

template = ChatPromptTemplate.from_template(prompt)
print(template)
```

```python
# {}안의 변수에 실제 값을 채워 넣으려면
template_message = template.format_messages(
    style = style,
    customer_email=customer_email
)

print(template_message)
```

```python
# 랭체인을 거쳐서 ChatGPT에 template_message를 전달
chat = ChatOpenAI(model='gpt-3.5-turbo', temperature=0.5)

response = chat(template_message)
print(response.content)
```

# 2. 파서(parser)

- AI 답변을 정해진 형태로 포맷팅

```python
customer_review = """\
This leaf blower is pretty amazing.  It has four settings:\
candle blower, gentle breeze, windy city, and tornado. \
It arrived in two days, just in time for my wife's \
anniversary present. \
I think my wife liked it so much she was speechless. \
So far I've been the only one using it, and I've been \
using it every other morning to clear the leaves on our lawn. \
It's slightly more expensive than the other leaf blowers \
out there, but I think it's worth it for the extra features.
"""

review_template = """\
For the following text, extract the following information:


gift: Was the item purchased as a gift for someone else? \
Answer True if yes, False if not or unknown.


delivery_days: How many days did it take for the product \
to arrive? If this information is not found, output -1.


price_value: Extract any sentences about the value or price,\
and output them as a comma separated Python list.


Format the output as JSON with the following keys:
gift
delivery_days
price_value


text: {text}
"""

{
    "gift":False,
    "delivery_days":5,
    "price_value":''
}
```

```python
prompt_template = ChatPromptTemplate.from_template(review_template)
print(prompt_template)
```

```python
message = prompt_template.format_messages(text=customer_review)
print(message)
```

```python
# chat = ChatOpenAI()
response = chat(message)
print(response.content)
```

```python
type(response.content)
```

```python
# response로 답변 받은 내용을 특정한 포맷(데이터 타입)으로 변화 -> Parser의 역할
from langchain.output_parsers import ResponseSchema     # '무엇'을 '어떤 형태'로 추출할까?
from langchain.output_parsers import StructuredOutputParser

# name -> 그래서 이렇게 추출한 애를 뭐라고 부를건데?
# description -> '무엇'을 '어떤 형태'로 추출할까?
gift_schema = ResponseSchema(name="gift",
                             description="Was the item purchased\
                             as a gift for someone else? \
                             Answer True if yes,\
                             False if not or unknown.")
delivery_days_schema = ResponseSchema(name="delivery_days",
                                      description="How many days\
                                      did it take for the product\
                                      to arrive? If this \
                                      information is not found,\
                                      output -1.")
price_value_schema = ResponseSchema(name="price_value",
                                    description="Extract any\
                                    sentences about the value or \
                                    price, and output them as a \
                                    comma separated Python list.")

response_schema = [gift_schema, delivery_days_schema, price_value_schema]

output_parser = StructuredOutputParser.from_response_schemas(response_schema)
```

```python
output_parser.get_format_instructions()
```

```python
review_template_2 = """\
For the following text, extract the following information:


gift: Was the item purchased as a gift for someone else? \
Answer True if yes, False if not or unknown.


delivery_days: How many days did it take for the product\
to arrive? If this information is not found, output -1.


price_value: Extract any sentences about the value or price,\
and output them as a comma separated Python list.


'\'문자가 답변에 포함되지 않아야 한다.


text: {text}


{format_instructions}
"""

# prompt 탬플릿을 활용함
prompt = ChatPromptTemplate.from_template(review_template_2)

# {} 안에 실제 값을 전달
instruction = output_parser.get_format_instructions()

# message로 합침
message = prompt.format_messages(text= customer_review,
                       format_instructions=instruction)

response = chat(message)
print(response.content)
```

```python
# 파싱
output = output_parser.parse(response.content)
print(output['gift'])
```
