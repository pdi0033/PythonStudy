```python
import os
from google.colab import userdata

os.environ['OPENAI_API_KEY'] = userdata.get('sesac7_openAI_key')
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate

# 체인 -> 랭체인 컴포넌트 중 하나
# 체인을 사용하면 각 작업을 연결하여 하나의 흐름을 만들 수 있음
from langchain.chains import LLMChain
from langchain.chains import ConversationChain

# 메모리
from langchain.memory import ConversationBufferMemory
from langchain.memory import ConversationBufferWindowMemory
from langchain.memory import ConversationTokenBufferMemory
from langchain.memory import ConversationSummaryMemory
```

```python
llm = ChatOpenAI()

# 대화를 체인으로 만듦 + 체인에 '메모리'를 붙여줌
memory = ConversationBufferMemory()

# 대화 체인 : 하나의 gpt 창을 계속 쓰면서 대화의 흐름을 유지할 수 있게
# 대화 흐름을 유지할 때, 기존의 기록에 대한 저장 용량 -> '메모리'
conversation = ConversationChain(
    llm= llm,  # llm=ChatOpenAI() # 이 채팅창에서 답변해줄 gpt 모델.
    memory= memory,     # 채팅창의 저장 용량
    verbose= True       # 채팅 기록 열람(자세히 보기)
)

conversation.predict(input='안녕, 나는 영등포 새싹에서 AI를 공부하고 있어.')
```

```python
conversation.predict(input='전세계적으로 일어나는 기후 재난은 뭐가 있어?')
```

```python
conversation.predict(input='내가 어디서 뭘 공부하고 있다고 했지?')
```

```python
# 윈도우를 기준으로 메모리를 셋업
memory = ConversationBufferWindowMemory(k=1)

conversation = ConversationChain(
    llm = ChatOpenAI(),
    memory = memory,
    verbose = True
)

conversation.predict(input='안녕, 나는 영등포 새싹에서 AI를 공부하고 있어.')
```

```python
conversation.predict(input='여름에 온열질환에 걸리지 않으려면 어떻게 해야 할까?')
```

```python
conversation.predict(input='내가 어디에서 뭘 공부하고 있다고 했지?')
```

# "토큰" 버퍼 메모리

'토큰' -> 자연어의 단위

토큰을 기준으로 LLM의 비용을 계산


```python
# 토큰화 라이브러리
!pip install tiktoken
```

```python
llm = ChatOpenAI(temperature=0, model='gpt-4o-mini')

# 토큰 버퍼 메모리 -> max_token_limit
memory = ConversationTokenBufferMemory(max_token_limit=50,
                                       llm=llm)
cv = ConversationChain(llm=llm,
                       memory=memory,
                       verbose=True )
```

```python
#memory에 대화내역을 (인위적으로) 삽입
memory.save_context({"input": "AI is what?!"},
                    {"output": "Amazing!"})
memory.save_context({"input": "Backpropagation is what?"},
                    {"output": "Beautiful!"})
memory.save_context({"input": "Chatbots are what?"},
                    {"output": "Charming!"})
```

```python
memory.load_memory_variables({})
```

```python
memory.save_context({"input":"Day and...."},
                    {"output":"What a beautiful Day!!!"})
```

```python
memory.load_memory_variables({})
```

# SummaryMemory

=> 요약 저장

```python
# create a long string
schedule = """오전 10시: 클라이언트와 온라인 미팅 (Zoom)
→ 새 프로젝트 요구사항 논의
→ 모델 구조 제안서 공유 예정이므로 미리 정리해 둘 것


오후 1시 ~ 5시: 미팅 내용 기반으로 모델 프로토타입 개발
→ 데이터 전처리 파이프라인 구성
→ Transformer 기반 모델 베이스라인 학습 돌려보기
→ 저녁 세미나 전에 체크포인트 저장 필수


오후 7시: 대학원 세미나 발표 (오프라인)
→ 발표 주제: "Diffusion 모델의 최신 동향 및 응용 사례"
→ 발표 자료와 예제 코드 노트북 준비
→ Q&A 대비해서 논문 내용 숙지해 둘 것


※ 오전-오후-저녁까지 일정 가득함, 집중력과 체력 분배 중요"""
```

```python
memory = ConversationSummaryMemory(llm=llm, max_token_limit=50)
```

```python
memory.save_context({'input':f'My Daily schedule is.. {schedule}'},
                    {'output': 'You are so busy...'})
```
