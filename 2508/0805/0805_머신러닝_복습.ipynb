```python
# 상관계수값이 너무 낮으면 제거해야 한다.
# 0.05보다 낮은 것들
XX = X
XX['Pass/Fail'] = y
print(XX.shape)

correlations = XX.corr()['Pass/Fail'].abs()
print(correlations)

# 기준점 잡기
threshold = 0.05
result_index = correlations[correlations >= threshold].index
print(len(result_index))

XX = XX[result_index]   # 너무 낮은 상관계수를 갖는 특성은 제거하자.
print(XX.shape)
```

```python
# 다중공선성 -> R언어는 제공한다.
# 각 속성 간에 너무 높은 관계성으로 인해서 제거해야할 속성이 있다.
# 두 개 이상의 특성이 상관관계가 너무 높을 때 이를 다중공선성이라고 한다.
# A -> B -> C -> D(마지막 목적지가 D라고 할 때)
# A, B, C나 사실상 같은 것. 나머지 특성을 제거해야 한다.
# 87 개의 특성에서 다중공선성문제를 해결하자
correlation_matrix = XX.corr()
print(correlation_matrix.shape)

# 타겟과 다른 feature 간의 상관관계를 계산해보자
# 상관계수 행렬은 각 특성들 간의 상관관계를 보여준다.
# 그리고 Target은 언제나 1이다.
# Pass/Fail을 제거한 다음에
target_correction = correlation_matrix['Pass/Fail'].drop('Pass/Fail')
print(correlation_matrix)

threshold = 0.8
# 상관계수의 임계점을 설정하자 서로간에 0.8 이상의 상관계수를 갖는 특성들을 중복으로 간주한다.
highly_correlated_features = set()  # 중복성을 제거하기 위해서 set타입을 사용한다.
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) >= threshold:
            colname_i = correlation_matrix.columns[i]
            colname_j = correlation_matrix.columns[j]
            highly_correlated_features.add(colname_i)
            highly_correlated_features.add(colname_j)
            # highly_correlated_features.add((colname_i, colname_j))

# 보존할 특성을 선택한다. - 다중공선성이 있는 특성들 중에서 하나만 고른다.
# 최대의 상관관계를 갖고 있는 특성을 가져온다.
features_to_keep = []
if not target_correction.empty:
    max_corr_feature = target_correction.idxmax()   # 해당 특성에서 가장 높은 상관관계를 찾는다.
    for feature in highly_correlated_features:      # 최대값과 같은 특성을 찾아서 저장한다.
        if feature == max_corr_feature:
            features_to_keep.append(max_corr_feature)

print(highly_correlated_features)
print("#######################################################3")
print(features_to_keep)

# 다중공선성이 높은 것들끼리 비교해서 중복되는 거 버리고,
# 상관계수가 낮은 것의 정보를 보관
non_correlated_features = set(XX.columns) - set(highly_correlated_features)
features_to_keep.extend(non_correlated_features)
print(features_to_keep)

# 중복배제
features_to_keep = list(set(features_to_keep))
print(len(features_to_keep))    # 36개 필드 features_to_keep: 사용해야할 특성에 대한 정보

X = XX[features_to_keep]
print(X.shape)


```

```python
# 표준화 정규화
# 정규화 - 데이터를 일정 범위(0~1) 사이에 머무르게,
# 표준화 - 데이터를 평균이 0 편차가 1인 정규분포로 전환한다.
# 정규화를 하는 이유
    # 서포트벡터머신, 딥러닝은 필수임.
    # 각 특성간의 단위의 크기를 비슷하게 맞추어서 특정 특성이 과도하게 중요한 요소로 인식되는 것을 막는다.
scaler = StandardScaler()
print(X.shape)
print(df.shape)
X = scaler.fit_transform(X)    # numpy 배열로 변환됨
df = pd.DataFrame(X)    # 데이터 프레임으로 만든다.
df['target'] = y
print(df)
print(df.describe())
```

```python
# 업샘플링 - 데이터를 증가시킨다. (언제나 좋은 결과를 가져오지는 않는다.)
# 데이터셋이 지나치게 불균형일 때. 작은쪽 데이터를 늘린다. (타켓층에서 얘기임)
# imblearn 라이브러리에 SMOTE 함수가 있다.
# 구성원리
# 1. 이웃찾기-knn 이웃알고리즘처럼.
# 2. 합성샘플을 생성한다. (이웃샘플하고 차이를 계산해서 그 차이를 임의의 비율로 확장해서 새로운 데이터를 생성한다.)
smote = SMOTE(sampling_strategy='minority', random_state=42)
# 소수 클래핑만을 대상으로 업샘플링을 한다.
X_up, y_up = smote.fit_resample(X, y)
print(X_up.shape)
print(y_up.shape)
print(y_up.value_counts())

```

```python
# 주성분 분석도 해보고, 고차원데이터를 => 저차원으로 변환한다.
# 데이터셋에서 중요한 정보는 그대로 보존하고 불필요하거나 상관관계가 높은 특성들을 결합하여 새로운 필드를 만들어 낸다.
from sklearn.decomposition import PCA
pca = PCA(n_components=24)  # 매개변수인 n_components가 우리가 추출하고자 하는 특성의 개수임.
                            # 이 수치는 마음대로 사용하는 거라 이것도 계속 바꿔가면서 테스트를 해봐야 안다.
                            # 주성분의 24개가 적당한지는 모른다. 그냥 한번 해보는 것.

# 이전에는 fit 호출하고 transform을 호출 => 최근에는 fit_transform 함수로 추가
X_up = pca.fit_transform(X_up)      # fit_transform 함수가 나중에 나오면서 내부적 처리가 추가되었음
print(X_up.shape)
# 주성분의 개수가 24개가 적당한지 확인하는 차트 그리기
# 각 주성분이 전체 데이터 분산 중 얼마나 많은 부분을 설명하는지 알아보기
cumlative_variance = np.cumsum(pca.explained_variance_ratio_)
print(cumlative_variance)
# np.cumsum: 비율들을 누적하여 새로운 배열을 만든다.
        # 주성분 개수가 추가되었을 때 분산의 변화를 설명한다.
```

```python
plt.figure(figsize=(10, 10))
plt.step(list(range(1, 25)), cumlative_variance, where='mid')   # 계단 차트
plt.ylabel('Cumlative variance explained')
plt.xlabel('Number of P A components')
plt.xticks(list(range(1,35)))
plt.axhline(0.91, color='r', linestyle='--', label='90% Variance')
plt.axvline(24, color='b', linestyle='--', label='24 components')
plt.legend()
plt.show()
```
